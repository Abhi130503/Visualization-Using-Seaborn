## Analysis of NEstaff in year 2019-2020 Vs 2020-2021

```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Since we have 2 separate files, These files are first concated into a single data frame for better comparison
import seaborn as sns
import pandas as pd

# import the first data frame
df1 = pd.read_csv('NEstaff_2019-20.csv')

# import the second data frame
df2 = pd.read_csv("NEstaff_2020-21.csv")

# concatenate the two data frames into a single data frame
df = pd.concat([df1, df2])#, keys=['Academic Year', 'Academic Year'])
df

```

```python
df.info()  # Checking the datatype and Not Null Values
```

```python
#Reading the data and finding unique values in each sub-categories :
Unique_HE=df['HE Provider'].unique()
print('HE Provider : \n' + Unique_HE)
Unique_AEF=df['Academic employment function'].unique()
print('Academic employment function : \n'+ Unique_AEF)
Unique_CL=df['Contract levels'].unique()
print('Contract levels : \n' + Unique_CL)
Unique_CS=df['Contract salary'].unique()
print('Contract salary : \n' + Unique_CS)
```

### Checking the overview of Female Employment in year "2019/20" for each catagory

```python
df_A = df[(df["Academic Year"] == "2019/20") & (df["Sex"] == "Female")]
df_A
```

### 1. Checking the over all Percentage distributionof Female in each category using Pie chart - 2019/20

```python
plt.figure(figsize=(25,25))                                                      #Setting the figure size

df_A = df[(df["Academic Year"] == "2019/20") & (df["Sex"] == "Female")]     #Applying filter to select only female and 2019/20

plt.subplot(5,1,1)                                                               #Subplot 1 
df_A1 = df_A.groupby(["HE Provider"]).Number.sum().reset_index()                 #Grouping values by HE Provider
plt.pie(data=df_A1, x='Number',labels = "HE Provider", autopct='%1.1f%%')        #Plotting the piechart
plt.legend(bbox_to_anchor=(1.2, 1.2), loc='upper left', borderaxespad=0)         #Adding legend
plt.title('Distribution of HE Provider for Female')                              #Adding title of subplot

plt.subplot(5,1,2)                                                                            #Subplot 2
df_A2 = df_A.groupby(["Academic employment function"]).Number.sum().reset_index()             #Grouping values by 
plt.pie(data=df_A2, x='Number',labels = 'Academic employment function', autopct='%1.1f%%')    #Academic employment function
plt.legend(bbox_to_anchor=(1.2, 1.2), loc='upper left', borderaxespad=0)
plt.title('Distribution of Academic employment function for Female')


plt.subplot(5,1,3)                                                                          #Subplot 3
df_A3 = df_A.groupby(["Contract salary"]).Number.sum().reset_index()                        #Grouping values by Contract salary
plt.pie(data=df_A3, x='Number',labels = "Contract salary", autopct='%1.1f%%')
plt.legend(bbox_to_anchor=(1.2, 1.2), loc='upper left', borderaxespad=0)
plt.title('Distribution of Contract salary for Female')

plt.subplot(5,1,4)                                                                          #Subplot 4
df_A4 = df_A.groupby(["Contract levels"]).Number.sum().reset_index()                        #Grouping values by Contract levels
plt.pie(data=df_A4, x='Number',labels = 'Contract levels', autopct='%1.1f%%')
plt.legend(bbox_to_anchor=(1.2, 1.2), loc='upper left', borderaxespad=0)
plt.title('Distribution of Contract levels for Female')


plt.show()


```

### 7. Heat Map Using Pivot table to Show the Number of Female in each category.

```python
#Step 1:Created a plot table using function : pivot_table()
#Step 2: Creating a Heat Map of the data using pivot table
#Step 3: Printing the HeatMap and Pivot table

plt.figure(figsize=(5,15))
pivot_table = df_FE.pivot_table(values=['Number'],index=["Academic Year",'HE Provider','Academic employment function','Contract levels'], aggfunc='sum', fill_value=0, margins=False)
sns.heatmap(pivot_table,annot=True,fmt='.0f',linecolor='black',linewidths=0.5,cmap='mako_r'
            ,annot_kws=dict(size=15,color='red',weight='bold'))
print(pivot_table)
```
